```python
from calculations.LOCA2_predagster import OpenLocaCat
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
query = dict(variable=['tasmax'],
             model=['FGOALS-g3'],
             scheme=['historical']
            )

OpenLocaCat = OpenLocaCat()

dataset = OpenLocaCat.load(query)

dataset
```

```python
plt.rcParams['figure.figsize'] = [10,7]
fig, ax = plt.subplots(1,1, subplot_kw={'projection':ccrs.PlateCarree(central_longitude=180)})
dataset.tasmax_tavg.sel(time='2000-07-16').mean('member_id').plot(ax=ax, transform=ccrs.PlateCarree())
ax.add_feature(cfeature.STATES)
```

```python
dataset_ill = OpenLocaCat.illinois(dataset)
dataset_ill
```

```python
plt.rcParams['figure.figsize'] = [10,7]
fig, ax = plt.subplots(1,1, subplot_kw={'projection':ccrs.PlateCarree(central_longitude=180)})
dataset_ill.tasmax_tavg.sel(time='2000-07-16').mean('member_id').plot(ax=ax, transform=ccrs.PlateCarree())
ax.add_feature(cfeature.STATES)
```

```python
dataset_stats2 = OpenLocaCat.stats(dataset_ill, ['time','member_id'])
dataset_stats2
```

```python
dataset_stats = OpenLocaCat.stats(dataset_ill, ['member_id'])
dataset_stats
```

```python
plt.rcParams['figure.figsize'] = [20,7]
fig, [ax1,ax2,ax3] = plt.subplots(1,3, subplot_kw={'projection':ccrs.PlateCarree(central_longitude=180)})

dataset_stats.tasmax_tavg.sel(time='2000-07-16').sel(stats='mean').plot(ax=ax1, transform=ccrs.PlateCarree())
ax1.add_feature(cfeature.STATES)

dataset_stats.tasmax_tavg.sel(time='2000-07-16').sel(stats='stdev').plot(ax=ax2, transform=ccrs.PlateCarree())
ax2.add_feature(cfeature.STATES)

dataset_stats.tasmax_tavg.sel(time='2000-07-16').sel(stats='variance').plot(ax=ax3, transform=ccrs.PlateCarree())
ax3.add_feature(cfeature.STATES)
```
